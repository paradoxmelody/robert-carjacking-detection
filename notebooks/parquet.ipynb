```python
#DATA REDUCTION

df = df.drop_duplicates()
# Remove a certain column from data
df.drop(columns=['eventid'], inplace=True)
df.drop(columns=['closest_suburb'], inplace=True)
df.info()
```

```python
#parse unitdatetime to int
df['unitdatetime'] = pd.to_datetime(df['unitdatetime'])
```

```python
#Time patterns
#when are vehicles most active?
df['hour'] = df['unitdatetime'].dt.hour
df['dayofweek'] = df['unitdatetime'].dt.day_name()

df['hour'].value_counts().sort_index().plot(kind='bar')
```

```python
#gforce exploration
sns.boxplot(data=df[['gforce_forward','gforce_backward','gforce_up','gforce_down','gforce_left','gforce_right']])
```

```python
#map the gforce
m = folium.Map(location=[df['latitude'].mean(), df['longitude'].mean()], zoom_start=11)

for _, row in df.sample(500).iterrows():
    folium.CircleMarker([row['latitude'], row['longitude']], radius=2, color='red').add_to(m)

m.save('map.html')
```

FLAG SUSPICIOUS STATIONARY EVENTS

```python
#FLAG SUSPICIOUS STATIOANRY EVENTS
df['is_stationary'] = df['speed'] == 0
df['hour'] = df['unitdatetime'].dt.hour
df['is_odd_hour'] = df['hour'].between(0, 3) | df['hour'].between(20, 23)

df['suspicious'] = df['is_stationary'] & df['is_odd_hour']
print(df['suspicious'].sum())   
```

```python
#FLAG G-FORCE ANOMALIES
df['gforce_anomaly'] = (
    (df['gforce_forward'] > 200) |
    (df['gforce_left'] > 200) |
    (df['gforce_right'] > 200) |
    (df['gforce_backward'] > 200)
)
print(df['gforce_anomaly'].sum())
```

```python
#COMBINED RISK FLAGGING
df['risk_event'] = df['suspicious'] | df['gforce_anomaly']
print(df['risk_event'].value_counts())
```

```python
df.to_csv('../data/parquet_cleaned.csv', index=False)
```
